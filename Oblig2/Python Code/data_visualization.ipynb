# Importing Oystein's recorded room temperatures

```python
import pandas as pd
from IPython.display import display
import datetime
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

file_df = pd.read_excel("OysteinsRoom-230210-2333.xlsx")

# Changing time format to ms since beginning.
start_time = file_df.iloc[0]['createAt']
start_time_dt = datetime.datetime.strptime(start_time, '%Y-%m-%d %H:%M:%S:%f')
print(f"The start time is:  {start_time_dt}")
for index, row in file_df.iterrows():
    current_time_dt = datetime.datetime.strptime(row['createAt'], '%Y-%m-%d %H:%M:%S:%f')
    time_delta = current_time_dt - start_time_dt
    min_since_start = time_delta.total_seconds()/60
    file_df.loc[index,'createAt'] = min_since_start
    
display(file_df)
```

```python
# Only plot temperature over time
fig_temp = px.line(
    data_frame = file_df,
    x="createAt",
    y="temperature",
    title='Oystein\'s recorded room temperatures',
    height=600,
    labels=dict(
        createAt="Time (minutes)",
        temperature="Temperature (°C)"
        )          
    )
fig_temp.show()
```

```python
#Plotting both subplots and on/off in same plot
plot_on_off = go.Scatter(
    x=file_df["createAt"],
    y=file_df["ON/OFF"],
    name="On/Off"
)

plot_temp = go.Scatter(
    x=file_df["createAt"],
    y=file_df["temperature"],
    name="Temperature (°C)"
)

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(plot_temp, secondary_y=False,)
fig.add_trace(plot_on_off,secondary_y=True)
fig['layout'].update(height = 600)
fig.show()
```
